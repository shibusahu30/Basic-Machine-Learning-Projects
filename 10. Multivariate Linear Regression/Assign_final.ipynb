```
def fillna(df): # Replacing Null Data
  list_obj = list(df.select_dtypes(include='object').columns)
  list_numric = list(df.select_dtypes(exclude='object').columns)

  for _ in list_obj:
      df[_].fillna(value=df[_].mode()[0], inplace=True) # Replace Object data type with its mode
  for _ in list_numric:
      df[_].fillna(value=df[_].mean(), inplace=True) # Replace numerical with mean
```

```
def one_hot_encode(df): # One hot encode the categorical data into numerical value
  list_obj = list(df.select_dtypes(include='object').columns)
  le = LabelEncoder()

  for _ in list_obj:
    df[_] = le.fit_transform(df[_])
```

## Chooose Variables

```
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1:].values
print(X.shape)
print(y.shape)
```

## Data Splitting

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Choosing Principal Component

```
pca = PCA(n_components = None)
train = pca.fit_transform(X_train)
test  = pca.transform(X_test)
explained_variance = pca.explained_variance_ratio_
print(explained_variance)
```

```
sum(explained_variance)
```

```
s = 0
i = 0
for _ in explained_variance:
  s += _
  print(i, s)
  i += 1
```

```
# Choosing 49 components gives variance accuracy of 0.999999964066085
pca = PCA(n_components = 49)
X_train = pca.fit_transform(X_train)
X_test  = pca.transform(X_test)
```

## Create Weight and Bias Vector

```
# creating input feayure vector with first column as 1
X = np.ones(shape=(X_train.shape[0], 1+X_train.shape[1]))
X[:, 1:] = X_train 
```

```
X.shape
```

## Calculating Optimized value for Bias and Weight Vector

```
# calculationg optimal weight and bias vector
inv = np.linalg.inv(np.dot(X.T, X))
theta = np.dot(inv, X.T)
theta = np.dot(theta, y_train).reshape(-1, 1)
```

```
print(theta.shape)
```

## Testing of Model

```
# input feature vector for testing purpose
X_testing = np.ones(shape=(X_test.shape[0], 1+X_test.shape[1]))
X_testing[:, 1:] = X_test 
```

## Model Prediction

```
# predicting output using linear regression
predict = np.dot(X_testing, theta).astype('int64')

for i in range(predict.shape[0]):
  print("Predicted:", predict[i], "Correct Label:", y_test[i])
```

## Error Computation

```
error = y_test - predict
sq_err = np.square(error)
RMSE = np.sqrt(np.sum(sq_err))
print("RMSE:", RMSE)
```
